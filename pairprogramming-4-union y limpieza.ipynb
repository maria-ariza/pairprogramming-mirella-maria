* ExplicaciÃ³n 

Hemos usado .merge para unir df1 y df2, ya que son dos dataframes con el mismo nÃºmero de filas pero distintas columnas, teniendo en comÃºn una de ellas. Para unirlos, aplicamos how='outer', esto quiere decir que el resultado es un dataframe con todo lo contenido en ambos dataframes, especificando las columnas por las que queremos unir, en este caso, 'Country' en df1 y 'country' en df2.

```python
df = df1.merge(df2, how='outer', left_on='Country', right_on='country')
```

1. DespuÃ©s de la uniÃ³n de datos, tenemos dos columnas de "country". Elimina una de ellas.

```python
df.drop('country', axis=1, inplace=True)
```

2. Los nombres de las columnas no son homogeneos. Cambia los nombres de las columnas de tal forma que:

No tengan espacios.

EstÃ©n en minÃºscula.

No tengan parÃ©ntesis, es decir, quitar "(%)", "(Km2)".

Algunas columnas tiene "\n". Eliminalos de los nombres de las columnas.

Algunas columnas tienen ":". Eliminalos de los nombres de las columnas.

```python
col_anterior = list(df.columns)
```

```python
col_anterior
```

```python
colum = []
for i in range(0, len(col_anterior)):
    try:
        col_anterior[i].replace(' ', '_')
    except:
        pass
    try:
        col_anterior = col_anterior[i].replace(':', '')
    except:
        pass
    try:
        col_anterior = col_anterior[i].replace('(%)', '')
    except:
        pass
    try:
        col_anterior = col_anterior[i].replace('(Km2)', '')
    except:
        pass
    try:
        col_anterior = col_anterior[i].replace('\n', '')
    except:
        pass
    try:
        col_anterior = col_anterior[i].strip()
    except:
        pass
    try:
        col_anterior = col_anterior[i].lower()
    except:
        pass
```
